```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
d1 = pd.read_csv("wine.csv")
d1
```

# Standardize the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(d1)
x_scaled = pd.DataFrame(data = x_scaled,columns = ['Type', 'Alcohol', 'Malic', 'Ash', 'Alcalinity', 'Magnesium', 'Phenols',
       'Flavanoids', 'Nonflavanoids', 'Proanthocyanins', 'Color', 'Hue','Dilution', 'Proline'])
x_scaled
```

# P.C.A

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=1)
pca1=PCA(n_components=2)
pca2=PCA(n_components=3)
data = pca.fit_transform(x_scaled)
data1 = pca1.fit_transform(x_scaled)
data2 = pca2.fit_transform(x_scaled)
pca_components=pd.DataFrame(data)
pca_components1=pd.DataFrame(data1)
pca_components2=pd.DataFrame(data2)
print(f'{pca_components}\n{pca_components1}\n{pca_components2}')
```

```python
# Create scatter plot for n_Components = 2
plt.figure(figsize=(8,6))
plt.scatter(data1[:,0], data2[:,1])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```

```python
# Create scatter plot for n_components = 3
from mpl_toolkits.mplot3d import Axes3D
# Create 3D scatter plot
fig = plt.figure(figsize=(15,13))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data2[:,0], data2[:,1], data2[:,2], c= 'r',s = 50)
# Change the view angle
ax.view_init(elev=20, azim=40)  
# change these values as needed
ax.set_xlabel('First Principal Component')
ax.set_ylabel('Second Principal Component')
ax.set_zlabel('Third Principal Component')
plt.show()
```

```python
var = pca.explained_variance_ratio_
var1 = pca1.explained_variance_ratio_
var2 = pca2.explained_variance_ratio_
print(f'{var}\n{var1}\n{var2}')
```

# Hierarchical Clustering

# K-means Clustering

```python
# Determine the optimal number of clusters using the elbow method
sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000,n_init = 10).fit(data)
    sse[k] = kmeans.inertia_

plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")

# calculate the differences between consecutive points(Calculate differences: We want to know how much the sum of squares (the “tightness” of our clusters) changes when we add another cluster. So, we subtract each sum of squares from the one before it. This is like figuring out how much taller each of your friends is than the next.)
differences = np.diff(list(sse.values()))

# calculate the second derivative(Now we want to see how much those differences change. This is like if you then figured out how much the differences in your friends’ heights changed.)
second_differences = np.diff(differences)

# find the elbow point where the second derivative is maximum
elbow_point = np.where(second_differences == max(second_differences))[0][0] + 2
# adding 2 because the index is 0-based and we're calculating the second derivative

plt.axvline(x=elbow_point, color='r', linestyle='--')
plt.show()

import os
os.environ["OMP_NUM_THREADS"] = '2'
```
